```python
def string_length(s):
    print("string_length called")
    return len(s)

def int_length(i):
    print("int_length called")
    return string_length(str(i))

def get_length_of_elements(t):
    funcs = []
    for x in t:
        if isinstance(x, int):
            funcs.append(int_length)
        elif isinstance(x, str):
            funcs.append(string_length)
    return LengthOfElementIteratorFactory(t, funcs)

class LengthOfElementIteratorFactory:
    def _init(self, t, funcs):  # Correctly define __init_ with parameters
        self.t = t
        self.funcs = funcs

    def _iter(self):  # Must be __iter, not _iter
        return LengthOfElementIterator(self)

class LengthOfElementIterator:
    def _init(self, factory):  # Correctly define __init_
        self.t = factory.t
        self.funcs = factory.funcs
        self.index = 0

    def _next(self):  # Must be __next, not _next
        if self.index >= len(self.t):
            raise StopIteration
        result = self.funcs[self.index](self.t[self.index])
        self.index += 1
        return result

    def _iter_(self):
        return self  # Make it an iterable

# Sample usage
x = (10, 20, "Good", "Example of functional programming", 30)
b =get_length_of_elements(x)

print("\n\ncalled length of elements\n\n")
for i in b:
    print(i)
```
