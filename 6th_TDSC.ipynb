```python
fig=plt.figure(facecolor='lightblue')
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.45,0.05,0.5,0.4])
x=np.linspace(-5,5,100)
ax1.plot(x,x**2,label='square',color='purple')
ax1.plot(x,x**3,label='cube',color='pink')
ax1.legend()

#sin wave
ax2.plot(x, np.sin(x), color='purple', label='sine')
ax2.plot(x, np.cos(x), color='pink', label='cos')

ax2.set_title("Sine and Cos Wave")
ax2.legend()
```

```python
font_dict={}
```

```python
font_dict ={'color':'white','fontfamily':'monospace', 'fontweight':'extra bold'}

fig = plt.figure(facecolor="blue")
ax = fig.add_axes([0,0,1,1], facecolor="black")
x = np.linspace(-5,5,10)

ax.plot(x,x**2,'r--o',label="square")
ax.plot(x,x**3,'w:s',label="cube")
ax.legend()
ax.set_title("Anatomy of figure " ,fontsize =30 ,fontdict=font_dict)
ax.set_xlabel("x-axis",fontsize=20,fontdict=font_dict)
ax.set_ylabel("y-axis",fontsize=20,fontdict=font_dict)

ax.set_xticks(np.arange(-6,6))
ax.set_xticklabels([chr(65+i) for i in range(12)],rotation = 30 , fontdict=font_dict)

ax.set_yticks(np.arange(-130,130,20))
```
