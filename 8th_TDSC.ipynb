```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create the figure and axes
fig, ax = plt.subplots(figsize=(9, 6))

# Set the dimensions of the flag
flag_width = 9
flag_height = 6

# Colors in order: blue, yellow, red, white, green
colors = ['#003f87', '#fcd116', '#d62828', '#ffffff', '#007a3d']

# Create polygons for each color band
# Each band is a triangle extending from the bottom-left to the top-right

# Blue band
blue = patches.Polygon([[0, 0], [0, flag_height], [flag_width * 0.4, flag_height]], closed=True, color=colors[0])

# Yellow band
yellow = patches.Polygon([[0, 0], [flag_width * 0.4, flag_height], [flag_width * 0.65, flag_height], [0, 0]], closed=True, color=colors[1])

# Red band
red = patches.Polygon([[0, 0], [flag_width * 0.65, flag_height], [flag_width * 0.9, flag_height], [0, 0]], closed=True, color=colors[2])

# White band
white = patches.Polygon([[0, 0], [flag_width * 0.9, flag_height], [flag_width, flag_height], [flag_width, flag_height * 0.8], [0, 0]], closed=True, color=colors[3])

# Green band
green = patches.Polygon([[0, 0], [flag_width, flag_height * 0.8], [flag_width, 0]], closed=True, color=colors[4])

# Add patches to the axes
for patch in [blue, yellow, red, white, green]:
    ax.add_patch(patch)

# Set limits and remove axes
ax.set_xlim(0, flag_width)
ax.set_ylim(0, flag_height)
ax.axis('off')

# Show the flag
plt.title("Flag of Seychelles")
plt.show()
```
